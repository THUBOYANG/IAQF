```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
result = pd.read_csv('predict.csv')
print(result.tail(1))
print(len(result))
print(result.loc[:575])
real_price = result['label']
predicted_price = result['predict']
low_freq = result['low_freq']
```

```python
plt.plot(real_price, color = 'black', label = 'Real Credit Spread')
plt.plot(predicted_price, color = 'green', label = 'Predicted Credit Spread')
plt.title('Prediction')
plt.xlabel('Time')
plt.ylabel('Price')
plt.legend()
plt.show()
```
